# Context Manager

```python
with open("file.txt", mode="r") as f:
    content=f.read()
```

```python
print(content)
```

```python
print(f.read()) #won't read
```

```python
with open("file1.txt", "w") as f:
    f.write("hello python")
```

```python
with open("file1.txt","r") as f:
    print(f.read())
```
